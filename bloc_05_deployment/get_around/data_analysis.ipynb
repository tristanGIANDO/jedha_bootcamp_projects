```python
# Removing extreme outliers in mileage and price
pricing_cleaned = pricing_df[
    (pricing_df['mileage'] > 0) & 
    (pricing_df['rental_price_per_day'] > 0)
]

price_dist_fig = px.histogram(
    pricing_cleaned, 
    x='rental_price_per_day', 
    title='Distribution of Rental Prices',
    nbins=30, 
    labels={'rental_price_per_day': 'Rental Price per Day'},
    color_discrete_sequence=['#636EFA']
)

mileage_price_fig = px.scatter(
    pricing_cleaned, 
    x='mileage', 
    y='rental_price_per_day', 
    title='Rental Price vs Mileage',
    labels={'mileage': 'Mileage', 'rental_price_per_day': 'Rental Price per Day'},
    color_discrete_sequence=['#00CC96']
)

price_dist_fig.show()
mileage_price_fig.show()
```

# Delay Dataset

```python
delay_df = pd.read_excel("src/get_around_delay_analysis.xlsx")
```

```python
display(delay_df.head())
```

```python
delay_df["state"].value_counts()
```

```python
# Filter ended rentals and summarize delay statistics
ended_rentals = delay_df[delay_df["state"] == "ended"]
display(ended_rentals["delay_at_checkout_in_minutes"].describe())

# Count rentals with delays and without delays
display(ended_rentals["delay_at_checkout_in_minutes"].notnull().value_counts())
```

```python
delay_cleaned = ended_rentals[
    (ended_rentals['delay_at_checkout_in_minutes'] > -5000) & 
    (ended_rentals['delay_at_checkout_in_minutes'] < 5000)
]

delay_dist_fig = px.histogram(
    delay_cleaned, 
    x='delay_at_checkout_in_minutes', 
    title='Distribution of Checkout Delays',
    nbins=50, 
    labels={'delay_at_checkout_in_minutes': 'Delay at Checkout (Minutes)'}
)

delay_dist_fig.show()
```

```python
# Define thresholds for minimum delay in minutes (1 hour, 2 hours, 3 hours)
thresholds = [60, 120, 180]

# Calculate number of rentals affected for each threshold and type of check-in
affected_rentals = []
for threshold in thresholds:
    affected = delay_cleaned[
        (delay_cleaned['time_delta_with_previous_rental_in_minutes'] < threshold) &
        (delay_cleaned['time_delta_with_previous_rental_in_minutes'] > 0)
    ]
    affected_counts = affected['checkin_type'].value_counts().reset_index()
    affected_counts.columns = ['checkin_type', 'count']
    affected_counts['threshold'] = threshold
    affected_rentals.append(affected_counts)

# Concatenate all affected counts into a single DataFrame
affected_rentals_df = pd.concat(affected_rentals, ignore_index=True)

# Visualization: Number of rentals affected by threshold and check-in type
affected_rentals_fig = px.bar(
    affected_rentals_df,
    x='threshold',
    y='count',
    color='checkin_type',
    barmode='group',
    title='Number of Rentals Affected by Minimum Delay Threshold and Check-In Type',
    labels={
        'threshold': 'Minimum Delay Threshold (Minutes)',
        'count': 'Number of Rentals Affected',
        'checkin_type': 'Check-In Type'
    },
    color_discrete_sequence=px.colors.qualitative.Set1
)

# Show the visualization
affected_rentals_fig.show()
```

### Conclusions

* Mean delay ~60.7 minutes whereas median delay ~9 minutes.
* Minimum and maximum values may be outliers or errors.

```python
# Filter positive delays (only actual delays)
positive_delays = delay_cleaned[delay_cleaned['delay_at_checkout_in_minutes'] > 0]

# Define thresholds to analyze (in minutes)
delay_thresholds = [30, 60, 120]

# Count the number of delays greater than each threshold
delay_counts = []
for threshold in delay_thresholds:
    count = positive_delays[positive_delays['delay_at_checkout_in_minutes'] > threshold].shape[0]
    delay_counts.append({'threshold': threshold, 'count': count})

# Convert results to DataFrame
delay_counts_df = pd.DataFrame(delay_counts)

# Visualization: Number of rentals delayed beyond thresholds
delay_threshold_fig = px.bar(
    delay_counts_df,
    x='threshold',
    y='count',
    title='Number of Rentals Delayed Beyond Thresholds',
    labels={
        'threshold': 'Delay Threshold (Minutes)',
        'count': 'Number of Rentals'
    }
)

# Show the visualization
delay_threshold_fig.show()
```
