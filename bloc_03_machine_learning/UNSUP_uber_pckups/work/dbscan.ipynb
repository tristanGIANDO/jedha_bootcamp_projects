```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots as sp
```

```python
IDX = 0
```

```python
df = pd.read_csv(f"data/dataset_day_{IDX}.csv")
X = np.load(f"data/preprocessing_day_{IDX}.npy")
```

## DBSCAN

```python
0.5/100
```

epsilon radius = 0.5km

```python
db = DBSCAN(eps=.5, min_samples=10, metric="manhattan")
db.fit(X)

# remove outliers
df["cluster"] = db.labels_
df_filtered = df[df["cluster"] > -1]

# display
fig = px.scatter_mapbox(df_filtered, lat="Lat", lon="Lon", color="cluster",
                        mapbox_style="carto-positron")
fig.show()
```
